```python
#INSTALAR

# Usaremos o selenium precisamos baixar o 'webdriver' > basta ver no site ofícial
# baixar webdriver = chromedriver

#IMPORTAR
from selenium import webdriver
import pandas as pd
import unicodedata
```

```python
#PASSO 1 - ENTRAR NA INTERNET

#CRIANDO NAVEGADOR

navegador = webdriver.Chrome()
navegador.get('https://www.google.com.br/?hl=pt-BR')
```

```python
#PASSO 2 - IMPORTAR A BASE DE DADOS

dados = pd.read_excel('commodities.xlsx')
display(dados)
```

```python

#PASSO 3 - PARA CADA PRODUTO DA NOSSA BASE

#COMANDOS 

# .click() -> clicar
# .send_keys("texto") -> escrever
# .get_attribute('') -> pegar um valor

print("Calculado...")
for linha in dados.index:

    #PARA ENTRAR NO SITE 'MELHOR CAMBIO' - 'https://www.melhorcambio.com/produtos-hoje'
       
    #agora irei normalizar o produo de duas formas PARA CARACTERES ESPECIAIS
    produto = dados.loc[linha,'Produto']
    lin = f'https://www.melhorcambio.com/{produto}-hoje'

    # - 1º forma retirar caracter especial

    #produto=produto.replace('ó','o').replace('é','e').replace('ã','a').replace('ú','u').replace('á','a').replace('ç','c')

    # - 2º forma para qualquer caracter especial
    link = unicodedata.normalize("NFKD", lin).encode(encoding="ascii",errors="ignore")
    
    #saída será essa>>> b'https://www.melhorcambio.com/produtos-hoje'
    #proxímo passo retirar " b'...' "
    #mudar o tipo do link para 'STRING'
    link = str(link)
    link = link.replace("b'",'').replace("'",'')
    
    #para buscar no navegador
    navegador.get(link)
    
#PASSO 4 - PEGAR O PREÇO ATUAL DO PRODUTO
  
    preco = navegador.find_element('xpath', '//*[@id="comercial"]').get_attribute('value')
    
    #saída será essa > '23.322,333' isso é uma 'string' 
    #irei mudar esse texo para numero de ponto flutuante
    
    preco = preco.replace('.','').replace(',','.')
    preco=float(preco)
    
#PASSO 5 - ATUALIZAR O PREÇO NA BASE DE DADOS
    dados.loc[linha,'Preço Atual'] = preco

#exibir dados atualizados
display(dados)  
```

```python
#PASSO 6 - DECIDIR QUAIS PRODUTOS COMPRAR
dados['Comprar'] = dados['Preço Atual'] < dados['Preço Ideal']
display(dados)
```

```python
#PASSO 7 FECHAR NAVEGADOR
navegador.quit()
```

```python
#PASSO 8 - EXPORTAR A BASE DE DADOS ATUALIZADA

#caso não existe irar criar uma nova tabela
#'index=False'> retira a coluna de index
dados.to_excel('commodities_atualizado.xlsx',index=False)
print('Finalizou')
```
